##### import libraries

```python
import pandas as pd
import numpy as np

#natural language processing
#pip install nltk
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer

import re
```

# (2) Natural Language Processing

```python
#import .csv file
cleaned_df = pd.read_csv("cleaned_script_data.csv", index_col=0)

#reset index for cleaned_df
cleaned_df.reset_index(drop=True, inplace=True)

print(cleaned_df.shape)
cleaned_df.head()
```

```python
print(cleaned_df['speaker'].value_counts())
```

```python
#create 'text' variable of every script line
cleaned_text = pd.DataFrame(cleaned_df['line_text'])

#lets clean the first question as a first step
cleaned_text['line_text'][0]
```

```python
#only keep letters and replace other symbols with a white space in the first blurb
line = re.sub('[^a-zA-Z]', ' ', cleaned_text['line_text'][0])

#change letters to lower-case
line = line.lower()   

#split the first blurb which is a string into list
line = line.split()

#for loop to remove stop-words and to do lemmatization
wn = WordNetLemmatizer()
line = [wn.lemmatize(word) for word in line if not word in set(stopwords.words('english'))]

#join blurb back into a string from a list
line = ' '.join(line)

line
```

```python
print(len(pd.DataFrame(cleaned_df['line_text'])))
length = len(pd.DataFrame(cleaned_df['line_text']))
```

```python
corpus = []
for i in range(0, length):
    #only keep letters and replace other symbols with a white space in the first blurb
    line = re.sub('[^a-zA-Z]', ' ', cleaned_text['line_text'][i])

    #change letters to lower-case
    line = line.lower()   
    #split
    line = line.split()
    #for loop to remove stop-words and to do stemming
    wn = WordNetLemmatizer()
    line = [wn.lemmatize(word) for word in line if not word in set(stopwords.words('english'))]
    #join blurb back into a string from a list
    line = ' '.join(line)
    corpus.append(line)
```

```python
#write corpus array into a dataframe and save it to a csv
corpus_df = pd.DataFrame(corpus, columns = ['line_text'])
corpus_df.head()
```

```python
#write corpus_df to a .csv
corpus_df.to_csv('corpus_script_data.csv')
```
