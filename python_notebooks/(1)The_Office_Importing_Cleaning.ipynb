### There were numerous misspellings of the character names, so I decided to revise the character name data and include the totality of the lines. It took some time.

```python
options = ['Michael',
           'M ichael', 
           'MIchael',
           'Micael', 
           'Micahel', 
           'Michae', 
           'Michael ',
           'Michael [on phone]',
           'Michal', 
           'Micheal', 
           'Michel', 
           'Mihael', 
           'Miichael',
           'Dwight',
           'Dwight.', 
           'Dwight:', 
           'DwightKSchrute',
           'Dight',
           'Jim',
           'JIm',
           'Jim ',
           'Pam',
           'Pam ',
           'Pam [on phone]',
           'Ryan', 
           'Ryan Howard',
           'Andy', 
           'Robert', 
           'Stanley', 
           'Kevin', 
           'Meredith',
           'Angela',
           'Angels', 
           'Anglea',
           'Angel',
           'Oscar', 
           'Phyllis',
           'Phylis', 
           'Phyliss',
           'Bob Vance',
           'Bob',
           'Roy', 
           'Jan', 
           'Jan [on phone]'
           'Kelly', 
           'Toby', 
           'Creed',
           'Darryl',
           'Darrly', 
           'Darry', 
           'Daryl',
           'Erin',
           'Carol Stills', 
           'Carrol', 
           'Carroll',
           'Gabe', 
           'Holly',
           'Helen', 
           'Helene',
           'Nellie', 
           'Clark', 
           'Pete',
           'David',
           'Dacvid Walalce', 
           'Dacvid Wallace',
           'David Wallace',
           'David Wallcve', 
           'DeAgnelo', 
           'DeAngelo',
           'Deangelo',
           'Louanne',
           'Jo', 
           'Cathy', 
           'Devon',
           'Danny', 
           'A.J.'
           'AJ',
           'Karen', 
           'Rolando',
           'Stephanie',
           'Ben',
           'Charles'] 
  
# selecting rows based on condition 
cleaned_df = raw_script_df[raw_script_df['speaker'].isin(options)] 
  
print('\nResult dataframe :\n', cleaned_df.shape, '\n \n approximately 5k lines were removed - we were able to retain the majority of the data') 
```

```python
#replace all mistaken versions of character spellings with correct spellings

#Michael
cleaned_df.speaker.replace(['Michael'], ['Michael'], inplace=True)
cleaned_df.speaker.replace(['M ichael'], ['Michael'], inplace=True)
cleaned_df.speaker.replace(['Micael'], ['Michael'], inplace=True)
cleaned_df.speaker.replace(['Micahel'], ['Michael'], inplace=True)
cleaned_df.speaker.replace(['Michae'], ['Michael'], inplace=True)
cleaned_df.speaker.replace(['Michael '], ['Michael'], inplace=True)
cleaned_df.speaker.replace(['Michael [on phone]'], ['Michael'], inplace=True)
cleaned_df.speaker.replace(['Michal'], ['Michael'], inplace=True)
cleaned_df.speaker.replace(['Micheal'], ['Michael'], inplace=True)
cleaned_df.speaker.replace(['Michel'], ['Michael'], inplace=True)
cleaned_df.speaker.replace(['Mihael'], ['Michael'], inplace=True)
cleaned_df.speaker.replace(['Miichael'], ['Michael'], inplace=True)
cleaned_df.speaker.replace(['MIchael'], ['Michael'], inplace=True)

print('# of Michael')
print(cleaned_df.loc[cleaned_df['speaker']=='Michael','speaker'].agg(['count']))
```

```python
#Dwight
cleaned_df.speaker.replace(['Dwight'], ['Dwight'], inplace=True)
cleaned_df.speaker.replace(['Dwight.'], ['Dwight'], inplace=True)
cleaned_df.speaker.replace(['Dwight:'], ['Dwight'], inplace=True)
cleaned_df.speaker.replace(['DwightKSchrute'], ['Dwight'], inplace=True)
cleaned_df.speaker.replace(['Dight'], ['Dwight'], inplace=True)

print('# of Dwight')
print(cleaned_df.loc[cleaned_df['speaker']=='Dwight','speaker'].agg(['count']))
```

```python
cleaned_df.speaker.replace(['JIm'], ['Jim'], inplace=True)
cleaned_df.speaker.replace(['Jim '], ['Jim'], inplace=True)

cleaned_df.speaker.replace(['Pam '], ['Pam'], inplace=True)
cleaned_df.speaker.replace(['Pam [on phone]'], ['Pam'], inplace=True)
            
cleaned_df.speaker.replace(['Ryan Howard'], ['Ryan'], inplace=True)

cleaned_df.speaker.replace(['Angels'], ['Angela'], inplace=True) 
cleaned_df.speaker.replace(['Anglea'], ['Angela'], inplace=True)
cleaned_df.speaker.replace(['Angel'], ['Angela'], inplace=True)

cleaned_df.speaker.replace(['Phylis'], ['Phyllis'], inplace=True)
cleaned_df.speaker.replace(['Phyliss'], ['Phyllis'], inplace=True)

cleaned_df.speaker.replace(['Bob Vance'], ['Bob'], inplace=True)

cleaned_df.speaker.replace(['Jan [on phone]'], ['Jan'], inplace=True)

cleaned_df.speaker.replace(['Darrly'], ['Darryl'], inplace=True)
cleaned_df.speaker.replace(['Darry'], ['Darryl'], inplace=True)
cleaned_df.speaker.replace(['Daryl'], ['Darryl'], inplace=True)

cleaned_df.speaker.replace(['Carol Stills'], ['Carrol'], inplace=True)
cleaned_df.speaker.replace(['Carroll'], ['Carrol'], inplace=True)

cleaned_df.speaker.replace(['Helene'], ['Helen'], inplace=True)
    
cleaned_df.speaker.replace(['Dacvid Walalce'], ['David'], inplace=True)
cleaned_df.speaker.replace(['Dacvid Wallace'], ['David'], inplace=True)
cleaned_df.speaker.replace(['David Wallace'], ['David'], inplace=True)
cleaned_df.speaker.replace(['David Wallcve'], ['David'], inplace=True)

cleaned_df.speaker.replace(['DeAgnelo'], ['DeAngelo'], inplace=True)
cleaned_df.speaker.replace(['Deangelo'], ['DeAngelo'], inplace=True)

cleaned_df.speaker.replace(['AJ'], ['A.J.'], inplace=True)
```
